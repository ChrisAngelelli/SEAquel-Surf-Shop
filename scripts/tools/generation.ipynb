```python
import numpy as np
import pandas as pd
import random
```

```python
#configure idenification 

#surf board body
id_body = [*range(500001,500018)]
#resin to coat the body
id_resin = [*range(500018,500025)]
#adds high performace construction 
id_fiberglass = [*range(500025,500029)]
#construction of body
id_tap = [*range(500029,500035)]
#complementary leach
id_leach = [*range(500035,500036)]
#plugs  for fin and leach
id_plug = [*range(500036, 500043)]
#colors for broards
id_pigment = [*range(500043,500052)]
```

```python
dfd = [id_body, id_resin, id_fiberglass, id_tap, id_leach, id_plug, id_pigment]
```

```python
body_list = []
material_id = pd.DataFrame([])


for i in range(250):

    body_value = random.choice(id_body)
    resin_value = random.choice(id_resin)
    fiberglass_value = random.choice(id_fiberglass)
    tap_value = random.choice(id_tap)
    leach_value = random.choice(id_leach)
    plug_value = random.choice(id_plug)
    pigment_value = random.choice(id_pigment)

    body_list = [body_value,resin_value,fiberglass_value,tap_value,leach_value,plug_value,pigment_value]

 
    material_id = material_id.append(body_list)
#body_list = body_list.append(body_value)
```
